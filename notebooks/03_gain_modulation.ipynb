```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

%load_ext autoreload
%autoreload 2

sys.path.append(os.path.abspath(os.path.join('..')))
from src.theory import siegert_firing_rate

mu_values = np.linspace(0.01, 0.1, 400)
d_mu = mu_values[1] - mu_values[0]

sigma_values = [0.05, 0.15, 0.35]

plt.figure(figsize=(9, 6))
colors = ['darkorange', 'dodgerblue', 'crimson']

for sigma, color in zip(sigma_values, colors):
    
    rates = np.array([siegert_firing_rate(m, sigma) for m in mu_values])
    
    # derivative of the firing rate with respect to mu
    gain = np.gradient(rates, d_mu)
    
    plt.plot(mu_values, gain, linewidth=2.5, color=color, label=f'$\sigma$ = {sigma}')

plt.axvline(x=0.05, color='gray', linestyle='--', alpha=0.8, label='Critical Threshold ($\mu = 0.05$)')
plt.title("Correct Gain Modulation: Neuron Sensitivity", fontsize=15)
plt.xlabel("Mean Input Current ($\mu$)", fontsize=13)
plt.ylabel("Gain (df/d$\mu$)", fontsize=13)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```
